```python
def cut(line, distance):
    # Cuts a line in two at a distance from its starting point
    if distance <= 0.0 or distance >= line.length:
        return [geometry.LineString(line)]
    coords = list(line.coords)
    for i, p in enumerate(coords):
        pd = line.project(geometry.Point(p))
        if pd == distance:
            return [
                geometry.LineString(coords[:i + 1]),
                geometry.LineString(coords[i:])]
        if pd > distance:
            cp = line.interpolate(distance)
            return [
                geometry.LineString(coords[:i] + [(cp.x, cp.y)]),
                geometry.LineString([(cp.x, cp.y)] + coords[i:])]
```

```python
def cut_wind(wind_start, wind_end, line_start, line_end, geo, length):
    try:
        # Window is full line
        if (wind_start == line_start) and (wind_end == line_end):
            return geo
        # Window starts at beginning of line
        elif (wind_start == line_start) and (wind_end != line_end):
            dist = (wind_end - wind_start) / (line_end - line_start)
            new_geo = cut(geo, (dist * length))[0]
            return new_geo
        # Window starts middle of line and ends before end of line
        elif (wind_start != line_start) and (wind_end != line_end):
            dist = (wind_end - line_start) / (line_end - line_start)
            new_geo = cut(geo, (dist * length))[0]
            dist = (wind_start - line_start) / (line_end - line_start)
            new_geo = cut(new_geo, (dist * length))[1]
            return new_geo
        # Window starts middle ends at end of segment
        elif (wind_start != line_start) and (wind_end == line_end):
            dist = (wind_start - line_start) / (line_end - line_start)
            new_geo = cut(geo, (dist * length))[1]
            return new_geo
        else:
            print("failed out")
    except:
        return geo
```
